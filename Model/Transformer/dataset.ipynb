```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm

def convert_csv_to_npy(csv_path, output_folder, new_csv_path):
    os.makedirs(output_folder, exist_ok=True)

    df = pd.read_csv(csv_path)

    # Defensive check for necessary columns
    if 'Image' not in df.columns or 'Emotion' not in df.columns:
        raise ValueError(f"'Image' or 'Emotion' column not found. Found columns: {df.columns.tolist()}")

    new_data = []

    for _, row in tqdm(df.iterrows(), total=len(df), desc="Processing landmarks"):
        image_name = row['Image']
        emotion = row['Emotion']

        # Extract landmark values (all except 'Image' and 'Emotion')
        landmark_values = row.drop(['Image', 'Emotion']).values.astype(np.float32)

        # Save to .npy file
        landmark_filename = os.path.splitext(image_name)[0] + ".npy"
        landmark_path = os.path.join(output_folder, landmark_filename)
        np.save(landmark_path, landmark_values)

        # Record new entry
        new_data.append({
            "Image": image_name,
            "Landmark": landmark_path,
            "Emotion": emotion
        })

    # Save new CSV
    new_df = pd.DataFrame(new_data)
    new_df.to_csv(new_csv_path, index=False)
    print(f"\n✅ Saved new CSV with paths to: {new_csv_path}")

# === Paths (edit if needed)
train_csv = r"C:/Users/koushika/Dropbox/PC/Downloads/facemesh_data_train.csv"
test_csv = r"C:/Users/koushika/Dropbox/PC/Downloads/facemesh_data_test.csv"
output_landmark_dir = r"C:/Users/koushika/Dropbox/PC/Downloads/landmark_npy"

new_train_csv = r"C:/Users/koushika/Dropbox/PC/Downloads/facemesh_data_train_landmarks.csv"
new_test_csv = r"C:/Users/koushika/Dropbox/PC/Downloads/facemesh_data_test_landmarks.csv"

# === Run
convert_csv_to_npy(train_csv, output_landmark_dir, new_train_csv)
convert_csv_to_npy(test_csv, output_landmark_dir, new_test_csv)
```
